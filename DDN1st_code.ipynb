```python
import os
import numpy as np
import pandas as pd
df = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df['Gender'] = df['Gender'].astype('category')
df['Gender'] = df['Gender'].cat.codes
```

```python
df.head()
```

```python
# removing user id not required

df = df.iloc[:,1:]
df.head()
```

```python
# checking for duplicate values
df.isnull().sum()
```

```python
# splitting the data into training and testing
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
y.head()
```

```python
# split data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 42, stratify = y)
```

```python
## Multilayer Perceptrom MODEL (DNN?ANN)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense
from keras import Sequential
```

```python
## Case 1 ; BUilding Model using feature scaling
```

```python
model = Sequential()
# first hidden layer
model.add(Dense(128,activation = 'relu' , input_dim = 3))
# Second hidden layer
model.add(Dense(128,activation = 'relu'))
# output
model.add(Dense(1,activation = 'sigmoid'))
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy',metrics = ['accuracy'])
history = model.fit(x_train, y_train, validation_data = (x_test, y_test), epochs = 100)
```

```python
## using feature selection
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc_train_x = sc.fit_transform(x_train)
sc_test_x = sc.transform(x_test)
```

```python
model1 = Sequential()
#first hidden layer
model1.add(Dense(128,activation='relu', input_dim = 3))
# second hidden layer
model1.add(Dense(128,activation ='relu'))
#output
model1.add(Dense(1,activation = 'sigmoid'))
model1.summary()
```

```python

model1.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
history = model1.fit(sc_train_x, y_train, validation_data = (sc_test_x, y_test), epochs=100)
```

```python
model1 = Sequential()
#first hidden layer
model1.add(Dense(128,activation='relu', input_dim = 3,kernel_initializer= 'he_normal'))
# second hidden layer
model1.add(Dense(128,activation ='relu',kernel_initializer='he_normal'))
#output
model1.add(Dense(1,activation = 'sigmoid'))
model1.summary()
```

```python
model1.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
history = model1.fit(sc_train_x, y_train, validation_data = (sc_test_x, y_test), epochs=100)
```

```python
model= Sequential()
# first hidden layer
model.add(Dense(128,activation = 'sigmoid', input_dim=3,kernel_initializer='glorot_normal'))
#second hidden layer
model.add(Dense(128,activation = 'sigmoid', kernel_initializer='glorot_normal'))
model.add(Dense(128,activation = 'sigmoid', kernel_initializer='glorot_normal'))
model.add(Dense(128,activation = 'sigmoid', kernel_initializer='glorot_normal'))
model.add(Dense(128,activation = 'sigmoid', kernel_initializer='glorot_normal'))
model.add(Dense(128,activation = 'sigmoid', kernel_initializer='glorot_normal'))
model.add(Dense(128,activation = 'sigmoid', kernel_initializer='glorot_normal'))
model.add(Dense(128,activation = 'sigmoid', kernel_initializer='glorot_normal'))
model.add(Dense(128,activation = 'sigmoid', kernel_initializer='glorot_normal'))
#output
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
history= model.fit(sc_train_x, y_train, validation_data = (sc_test_x, y_test), epochs=100)
```

```python
import matplotlib.pyplot as plt

```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
## Regularisation method
model3 = Sequential()
# first hidden layer
model3.add(Dense(128, activation='relu', input_dim=3,kernel_initializer='he_normal',
                 kernel_regularizer=keras.regularizers.l1(0.001)))
# 2nd hidden layer
model3.add(Dense(128, activation='relu', kernel_initializer='he_normal',
                 kernel_regularizer=keras.regularizers.l1(0.001)))
# output
model3.add(Dense(1, activation='sigmoid'))

model3.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
history1 = model3.fit(sc_train_x, y_train, validation_data = (sc_test_x, y_test), epochs=100)
```

```python
import matplotlib.pyplot as plt

plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

```python
plt.plot(history1.history['accuracy'])
plt.plot(history1.history['val_accuracy'])
```
