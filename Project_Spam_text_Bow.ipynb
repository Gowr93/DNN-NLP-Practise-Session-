```python
import os
import numpy as np
import pandas as pd

dataset  = pd.read_csv('/content/SPAM_text_message.csv')
dataset.head()
```

```python
dataset['Category'].value_counts()/len(dataset)*100
# imbalanced dataset, so handle the data
# if missing data found, dont impute it just drop it
```

```python
dataset.isnull().sum().sum()
```

```python
ham = dataset[dataset['Category']=='ham']
#dataset with ham with store in ham
spam = dataset[dataset['Category']=='spam']
# spam data will stored separately in spam
```

```python
print(ham.shape, spam.shape)
```

```python
ham
```

```python
spam
```

```python
# Balancing the dataset
# target minority class to increase
# minority is spam

spam = spam.sample(ham.shape[0], replace = True)
```

```python
ham.shape[0]
```

```python
data = pd.concat([ham,spam])
```

```python
data.head()
```

```python
data.shape
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data['Message'],data['Category'], test_size = 0.3, random_state = 42)
```

## Feature Extraction and Model building

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
classifier = Pipeline([('bow', CountVectorizer()),
                       ('rfmodel', RandomForestClassifier())])

# Pipeline
```

```python
classifier.fit(x_train, y_train)
```

```python
y_pred_train = classifier.predict(x_train)
y_pred_tet = classifier.predict(x_test)
```

```python
y_pred_tet
```

```python
# Evaluate the model
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_train, y_pred_train))
print("****************************************************")
print(confusion_matrix(y_test, y_pred_tet))
```

```python
print(classification_report(y_train, y_pred_train))
print("****************************************************")
print(classification_report(y_test, y_pred_tet))
```

```python
print("Training Accuracy :", accuracy_score(y_train, y_pred_train))
print("****************************************************")
print("Test Accuracy :",accuracy_score(y_test, y_pred_tet))
```
